```python
from scipy.interpolate import interp1d
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# read in data
df = pd.read_csv("Heatmap Trial.csv")
```

```python
df.head()
```

```python
list1 = df.avg_temp_year.values.tolist()
```

```python
list1
```

```python
# interpolating circle radius to given circle range

m = interp1d([1,max(list1)],[5,16])

circle_radius = m(list1)
```

```python
typelist = ['open-street-map','white-bg','carto-positron','carto-darkmatter','stamen-terrain','stamen-toner','stamen-watercolor']

for i in typelist:
    print(i)
    fig = px.density_mapbox(df, lat='Lat', lon='Long', radius=circle_radius,zoom=0,
                           mapbox_style=i)
    
    fig.show()
```

```python
df = pd.read_csv('Heatmap Trial.csv')

import plotly.express as px
fig = px.density_mapbox(df, lat='Lat', lon='Long', z='avg_temp_year', radius=10,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="stamen-terrain")
fig.show()
```
