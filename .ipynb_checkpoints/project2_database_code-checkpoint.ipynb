```python
import pandas as pd
import numpy as np
import flask
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
#read in our csvs and transfer the csvs to dataframes
co2_csv = "assets/data/co2_nvk.csv"

co2_df = pd.read_csv(co2_csv)

co2_df.head(15)
```

```python
us_plants_csv = "assets/data/us_plants_nvk.csv"

us_plants_df = pd.read_csv(us_plants_csv)

us_plants_df.head(15)
```

```python
#create engine
engine = create_engine('sqlite://', echo=False)
```

```python
#df -> database
co2_df.to_sql('co2_database',con=engine)
```

```python
us_plants_df.to_sql('us_plants_db', con=engine)
```

```python
engine.execute("SELECT * FROM co2_database").fetchall()
```

```python
engine.execute("SELECT * FROM us_plants_db").fetchall()
```
